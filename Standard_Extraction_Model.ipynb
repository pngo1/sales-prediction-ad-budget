# **Extract the standard base and version**

```python
prod_des = pd.read_csv('/content/product_descriptions.csv')
prod_des.head(100)
```

```python
prod_des['description'] = prod_des['description'].apply(lambda x: ' '.join(x.split()))
prod_des
```

```python
def extract_standards(description):
  pattern = r'\b(?:EN\sISO\s\d{4}(?::\d{4})?|EN\s\d+(?:-\d+)*(?:-\d+)*|EN\d+(?:-\d+)*)| \b(?:IEC\s\d+(?:-\d+)*\s*[IV]*|ASTM\s[\w/-]+)\b | \b(?:FSC\s+certified|OEKOTEX)\b'
  matches = re.findall(pattern, description)
  if matches:
    return ', '.join(matches)
  else:
    return None
```

```python
prod_des['standards'] = prod_des['description'].apply(extract_standards)
```

```python
prod_des
```

```python
count = prod_des['standards'].notna()
results = prod_des[count]
results
```

# **Try out a machine learning model**







```python
sample_data = pd.read_csv('/content/train_test_data.csv')
sample_data.head(100)
```

```python
filtered_df = sample_data[(sample_data['type'] == 'standard-certification') | (sample_data['type'] == 'standards-base')| (sample_data['type'] == 'standards-test') | (sample_data['type'] == 'standards-version')]

filtered_df.head(100)

```

# **Evaluation**

evaluation for regex can be done by comparing the predicted standards and actual ones. There can be false negatives and false positives and there should be a way to balance the two. We can give a description from the list or create one ourselves and input the actual standards manually and comapre to the predicted by the function.

```python
# first setting the real and preducted standards
truth = ["EN374", "EN455", "EN1186"]
# Predicted matches from the regular expression
predicted = ["EN374", "EN455", "EN1186"]
```

```python
# then calculating true psoitives, false positives and false negatives
true_positives = len(set(truth) & set(predicted))
false_positives = len(predicted) - true_positives
false_negatives = len(truth) - true_positives
```

```python
# Calculate precision and recall
# precision: The proportion of true positive matches out of all positive matches.
precision = true_positives / (true_positives + false_positives)
# recall: The proportion of true positive matches out of all actual positive instances in the truth.
recall = true_positives / (true_positives + false_negatives)
#accuracy: The proportion of correctly classified matches out of all instances.
accuracy = true_positives / len(predicted)
```

```python
# Combination of precidsion and recall is F1 score
f1_score = 2 * (precision * recall) / (precision + recall)
```

```python
# Print the accuracy metrics
print("Precision: {:.2f}%".format(precision * 100))
print("Recall: {:.2f}%".format(recall * 100))
print("Accuracy: {:.2f}%".format(accuracy * 100))
print("F1 Score: {:.2f}%".format(f1_score * 100))
```

# **Standard test**

**Standard test and test results extraction**  
Observations :  Standard tests are not mentioned in products description. Each product has many standard tests, which belong to the standard base of the product.


```python
import re
import pandas as pd
```

```python
#Extract all standard tests from all labels
all_labels = pd.read_csv('all_labels.csv')
standard_tests = all_labels[all_labels['type'] == 'standards-test']

#Define a function extract
def get_standardTests(standardBase):
    standardBase = standardBase.lower()
    standardBase = standardBase.replace(" ", "-")
    possibleTests = standard_tests[standard_tests['code'].str.startswith(f'{standardBase}')]
    possible_tests_list = [(row['code'], row['label']) for index, row in possibleTests.iterrows()]
    return possible_tests_list
```

The function attempts to get all tests belonging to the standard base of the products, results in a selection of approximately 10 standard tests. The label then can be chosen manually based on tests' results.

```python
# Create an empty DataFrame to store the results
new_df = pd.DataFrame(columns=['standard', 'predict_tests'])

#Demonstrate on y_pred
# Iterate over each row in product labels
matching_tests = []
for i in y_pred:
    # Find standard tests that match the product description
    matching_tests.append(get_standardTests(i))
    # Append the information to the new DataFrame

new_df['standard'] = y_pred
new_df['predict_tests']= matching_tests

print(new_df.head(10))
```
